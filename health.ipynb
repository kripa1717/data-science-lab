```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
age = [21,30,35,39,64]
bmi = [150,183,149,166,147]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(age, bmi, color='red', marker='o', linestyle='solid')

# add a title
plt.title("Health Chart According To BMI")

# add a label to the y-axisYears
plt.ylabel("bmi")
plt.xlabel("age")
plt.show()
```

```python
flight = sns.load_dataset('flights') # load flight dataset
print(flight)

flight_new = flight.groupby('year').sum(numeric_only=True).reset_index()
# line chart
sns.lineplot(x='year',y='passengers',data=flight_new)
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align"
)
```

```python
age = [21,30,35,39,64]
bmi = [150,183,149,166,147]

# plot bars with left x-coordinates [0, 1, 2, 3, 4], heights [num_oscars]
plt.bar(range(len(age)), bmi)

plt.title("age")     # add a title
plt.ylabel("bmi")   # label the y-axis

# label x-axis with movie names at bar centers
plt.xticks(range(len(age)), bmi)

plt.show()
```
